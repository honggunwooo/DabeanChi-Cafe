```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
import warnings

# --- 0. Google Drive 마운트 (저장/불러오기를 위해) ---
drive.mount('/content/drive')
drive_path = '/content/drive/MyDrive/XGBoost_Models/'
# 폴더가 없으면 생성
if not os.path.exists(drive_path):
    os.makedirs(drive_path)

X = coffee_flavor.drop(['desc_1', 'flavor', 'body', 'acid', 'aroma', 'name'], axis=1)
y = coffee_flavor['desc_1']

# 학습/테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
print(f"X_train data types:\n{X_train.dtypes}")


# --- 2. 레이블 인코딩 (타겟 y가 문자열이므로 필수) ---
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
print(f"\ny 레이블 인코딩 완료. 클래스 매핑: {dict(zip(le.classes_, le.transform(le.classes_)))}")


# --- 3. 특징 데이터(X) 전처리 (object 타입 컬럼을 category 타입으로 변환) ---
# 오류 해결을 위해 문자열 컬럼을 XGBoost가 처리할 수 있는 category 타입으로 변환
categorical_cols = X_train.select_dtypes(include=['object']).columns
for col in categorical_cols:
    X_train[col] = X_train[col].astype('category')
    # 테스트셋도 동일하게 변환하며, 새로운 범주는 NaN으로 처리
    X_test[col] = X_test[col].astype('category').cat.set_categories(
        X_train[col].cat.categories
    ).fillna(np.nan)
print("\nX 데이터 전처리 완료 (object -> category 변환).")

# --- 4. 모델 설정 및 CPU 학습 (에포크 100회) ---
model = XGBClassifier(
    objective='multi:softprob',
    n_estimators=10000,  # 에포크 100회
    learning_rate=0.0001,
    use_label_encoder=False,
    eval_metric='mlogloss',
    enable_categorical=True # 범주형 데이터 처리 활성화
)

eval_set = [(X_train, y_train_encoded), (X_test, y_test_encoded)]
print("\n모델 학습 시작 (CPU 사용)...")
model.fit(X_train, y_train_encoded, eval_set=eval_set, verbose=False)
print("모델 학습 완료.")


# --- 5. 모델 및 인코더 저장 ---
model_save_path = os.path.join(drive_path, 'xgb_coffee_model.joblib')
encoder_save_path = os.path.join(drive_path, 'label_encoder_coffee.joblib')
joblib.dump(model, model_save_path)
joblib.dump(le, encoder_save_path)
print(f"\n모델과 인코더가 구글 드라이브에 저장되었습니다.")


# --- 6. 예측 및 평가 ---
from sklearn.metrics import accuracy_score, classification_report
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test_encoded, y_pred)
print(f"\n정확도: {accuracy:.4f}")
print("분류 보고서:")
print(classification_report(y_test_encoded, y_pred))
```

```python
def predict_with_user_input_coffee(model, encoder, feature_cols):
    print("\n--- 새로운 커피 데이터 입력 ---")
    user_inputs = {}

    # 실제 데이터 타입에 맞게 입력받기 (여기서는 roast, origin, location이 문자열)
    for col in feature_cols:
        if col in ['roast', 'origin', 'location']:
            value = input(f"{col} (문자열) 입력: ")
            user_inputs[col] = [value]
        else: # 숫자형 데이터
            value = float(input(f"{col} (숫자) 입력: "))
            user_inputs[col] = [value]

    user_data = pd.DataFrame(user_inputs)

    # 입력 데이터를 학습 데이터와 동일한 'category' 타입으로 변환
    for col in feature_cols:
        if col in ['roast', 'origin', 'location']:
            user_data[col] = user_data[col].astype('category').cat.set_categories(
                X_train[col].cat.categories # 학습 시 사용한 X_train의 범주 사용
            )
            # 학습 데이터에 없던 새로운 범주는 NaN이 됩니다. XGBoost가 NaN을 처리합니다.

    prediction_numeric = model.predict(user_data)
    predicted_label = encoder.inverse_transform(prediction_numeric)

    print("\n--- 예측 결과 ---")
    print(f"예측된 클래스 (이름): {predicted_label}")

# 함수 실행 (X_train의 컬럼 이름들을 사용하여 실행)
predict_with_user_input_coffee(model, le, X_train.columns)
```

```python
import pandas as pd
import numpy as np
import joblib
from xgboost import XGBClassifier
from google.colab import drive
import os

# --- 1. Google Drive 마운트 및 경로 설정 ---
drive.mount('/content/drive')

drive_path = '/content/drive/MyDrive/'
model_filename = 'xgboost_classifier_model.joblib'
encoder_filename = 'label_encoder.joblib'

model_load_path = os.path.join(drive_path, model_filename)
encoder_load_path = os.path.join(drive_path, encoder_filename)

# --- 2. 모델 및 LabelEncoder 불러오기 ---
try:
    loaded_model = joblib.load(model_load_path)
    loaded_encoder = joblib.load(encoder_load_path)
    print("모델과 LabelEncoder를 성공적으로 불러왔습니다.")
except FileNotFoundError:
    print("오류: 모델 또는 인코더 파일이 지정된 경로에 없습니다.")
    exit()

# --- 3. 사용자 입력 및 예측 함수 ---

def predict_with_user_input(model, encoder):

    # !!! 중요: 실제 사용 중인 컬럼 이름 리스트로 수정하세요 !!!
    # 예시: ['name', 'roast', 'origin', 'location', 'rating', 'price']
    feature_columns = ['roast', 'origin', 'location'] # 예시 컬럼명

    print("\n--- 새로운 데이터 입력 ---")
    user_inputs = {}

    for col in feature_columns:
        # 데이터 타입에 따라 입력 방식 결정 (문자열 vs 숫자)
        # 이 부분은 실제 데이터의 dtype에 맞게 조정해야 합니다.
        if col in ['roast', 'origin', 'location']:
            value = input(f"{col} (문자열) 입력: ")
            user_inputs[col] = [value]
        else: # 나머지 숫자형 컬럼 (예: rating, price)
            try:
                value = float(input(f"{col} (숫자) 입력: "))
                user_inputs[col] = [value]
            except ValueError:
                print(f"잘못된 입력 ({col}). 숫자를 입력해 주세요.")
                return

    # 입력받은 값을 DataFrame 형태로 변환
    user_data = pd.DataFrame(user_inputs)

    # !!! 중요: 학습 시 적용했던 전처리 (category 변환)를 여기서도 적용 !!!
    for col in user_data.columns:
        if user_data[col].dtype == 'object':
             user_data[col] = user_data[col].astype('category')
             # 실제 학습 데이터셋의 범주 정보로 설정해야 정확한 예측이 가능합니다.
             # user_data[col] = user_data[col].cat.set_categories(loaded_model.feature_names_in_).fillna(np.nan)


    # 모델로 예측 수행
    prediction_numeric = model.predict(user_data)
    prediction_proba = model.predict_proba(user_data)

    # 숫자 예측값을 원래 레이블 문자열로 디코딩
    predicted_label = encoder.inverse_transform(prediction_numeric)

    print("\n--- 예측 결과 ---")
    print(f"입력된 데이터:\n{user_data}")
    print(f"예측된 클래스 (이름): {predicted_label}")
    print(f"예측 확률: {prediction_proba}")

# --- 4. 함수 실행 ---
predict_with_user_input(loaded_model, loaded_encoder)
```
